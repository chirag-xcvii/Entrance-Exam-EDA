##Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['Academic Subjects.English.Average GPA']]  # Independent variable (feature)
y = df['Total.Verbal']  # Dependent variable (target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.xlabel('Verbal GPA')
plt.ylabel('Verbal Scores')
plt.title('Linear Regression: Verbal GPA vs. Verbal Scores')
plt.legend()
plt.show()

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

r_squared = model.score(X_test, y_test)
print("R-squared:", r_squared)
```
